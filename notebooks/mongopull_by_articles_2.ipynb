```python
import urllib
```

```python
def joinurl(p):
    cursite, linksite = p
    try:
        if linksite[0] != '/':
            return linksite
        pr = urllib.parse.urlsplit(cursite)
        dpr = urllib.parse.urlsplit(linksite)
        return 'r/' +  urllib.parse.urljoin(pr.scheme + '://' +pr.netloc,  dpr.path)
    except:
        return linksite
```

```python
ldf['rdest'] = pd.Series(map(joinurl, zip(ldf.url, ldf.dest)))
```
