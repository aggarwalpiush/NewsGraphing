```python
import urllib
```

```python
def joinurl(p):
    cursite, linksite = p
    try:
        linksite = linksite.strip('"')
        if linksite[0] != '/' and linksite[0] != '#' and linksite[0:2] != '../':
            return linksite
        pr = urllib.parse.urlsplit(cursite)
        dpr = urllib.parse.urlsplit(linksite)
        # 'r/'
        return  urllib.parse.urljoin(pr.scheme + '://' +pr.netloc,  dpr.path)
    except:
        return linksite
    
def processurl(urlstr):
    pr = urllib.parse.urlsplit(urlstr)
    #print(pr)
    if pr.netloc.startswith('www.'):
        return pr.netloc[3:] + pr.path
    else:
        return pr.netloc + pr.path
```

```python
ldf['rsrc'] = pd.Series(map(processurl,ldf.url))
```

```python
ldf['rdest'] = pd.Series(map(lambda x : processurl(joinurl(x)), zip(ldf.url, ldf.dest)))
```
