```python
import json
import numpy as np
import sklearn
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
with open("metadata.json", "r") as jfile:
    data = json.load(jfile)
for key in data:
    data[key] = [x[0] for x in data[key]]
```

```python
data
```

```python
for site in data:
    if sum(data[site][0:3]) > 1.5:
        print(data[site])
```

```python
for f in [1, 2, 3]:
    clf = RandomForestClassifier()
    xs = []
    ys = []
    for site in data:
        if data[site][10] != f and data[site][9] != 1:
            xs.append(data[site][(f-1)*3:f*3])
            ys.append(data[site][9])
    #print(xs)
    clf.fit(np.asarray(xs), np.asarray(ys))
    xs = []
    ys = []
    for site in data:
        if data[site][10] == f and data[site][9] != 1:
            xs.append(data[site][(f-1)*3:f*3])
            ys.append(data[site][9])
    print(clf.predict(np.asarray(xs)))
    #print(ys)
    print(clf.score(xs, ys))
```
